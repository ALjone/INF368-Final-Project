```python
directory = '/content/INF368-Final-Project/results'

#Loop through each folder in the results folder, equivalent to each dataset
for folder, _, _ in os.walk(directory):
  #Read all the results for each folder
  bert = pd.read_csv(folder+"/bert_results.csv")
  lambada = pd.read_csv(folder+"/LAMBADA_results.csv")
  GanBert = pd.read_csv(folder+"/GanBert_results.csv")

  #Format the data in way that makes more sense
  df = pd.DataFrame(columns=["num_per_sample", "BERT", "GanBert", "LAMBADA"])
  for i, num_per_sample in enumerate([5, 10, 25, 50]):
    row = {"num_per_sample": num_per_sample, "BERT": bert.iloc[i, 2], "GanBert":GanBert.iloc[i, 2], "LAMBADA":lambada.iloc[i, 2]}
    df = df.append(row, ignore_index = True)

  #Save the results
  df.to_csv(folder+/"full_result.csv")

  #Make a plot and save it 
  ax = df.plot(x="num_per_sample", xlabel="samples per label", ylabel = "accuracy").
  fig = ax.get_figure()
  #Check that the folder exists
  if not os.path.exists('../images'):
      os.mkdir('../images')
  #We assume the folder name is the same as the one for the dataset
  fig.savefig('../images/'+folder+".png")
```
